```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+','',text)
    text = re.sub('<.*?>+',b'',text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w*','',text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

# Building LSTM Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding, SpatialDropout1D
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer

# Tokenizing and padding
tokenizer = Tokenizer(num_words=5000, lower=True)
tokenizer.fit_on_texts(x)
X = tokenizer.texts_to_sequences(x)
X = pad_sequences(X, maxlen=300)

# Splitting data for LSTM
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Building LSTM model
model = Sequential()
model.add(Embedding(5000, 128, input_length=X.shape[1]))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Training LSTM model
history = model.fit(X_train, Y_train, epochs=5, batch_size=64, validation_data=(X_test, Y_test), verbose=2)

# Evaluate the model
lstm_pred = model.predict(X_test)
lstm_pred = (lstm_pred > 0.5).astype(int)
print(classification_report(Y_test, lstm_pred))
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
import numpy as np

# Assuming you have already defined and trained your LSTM model as 'model'
# And you have your tokenizer as 'tokenizer'

def manual_testing_lstm(news):
    # Preprocess the input news article
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)

    # Apply the same preprocessing steps used for training
    new_def_test['text'] = new_def_test['text'].apply(wordopt)  # Make sure `wordopt` is defined for text preprocessing

    # Tokenize and pad the sequences
    new_x_test = tokenizer.texts_to_sequences(new_def_test['text'])
    new_x_test = pad_sequences(new_x_test, maxlen=300)  # Ensure maxlen matches training

    # Predict using the LSTM model
    lstm_pred = model.predict(new_x_test)
    lstm_pred = (lstm_pred > 0.5).astype(int)

    # Output the prediction
    prediction = "Not A Fake News" if lstm_pred[0][0] == 1 else "Fake News"
    print(f"LSTM Prediction: {prediction}")

# Example usage
news = input("Enter a news article for LSTM model testing: ")
manual_testing_lstm(news)
```
